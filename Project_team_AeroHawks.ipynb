```python
import pandas as pd
import numpy as np
from numpy import nan
```

```python
app = pd.read_csv('/content/drive/MyDrive/Project/playstore_apps.csv')
app.head(3)
```

```python
# remove the duplicates
app.drop_duplicates(keep=False, inplace=True)
app.head()
```

```python
# update column type
app['Last Updated']=pd.to_datetime(app['Last Updated'])
app.info()
app.head(20)
```

```python
app.describe()
```

```python
# find the unique value for each category column
print(app['Type'].unique())
print(app['Content Rating'].unique())
print(app['Genres'].unique())
print(app['Android Ver'].unique())
print(app['Category'].unique())

```

```python
# clean the 'category' column:
app.drop(app[app['Category']=='1.9'].index, inplace=True)

# clean the 'Type' column:
app.drop(app[app['Type']=='0'].index, inplace=True)
print(app['Type'].unique())
print(app['Category'].unique())
```

```python
app.info()
```

```python
# cleaning the null values
# replace null values with 0 for column rating
app['Rating'] = app['Rating'].fillna(0)
# replace null values with NAN for column Current VER
app['Current Ver'] =app['Current Ver'] .replace(" ",np.NaN)
# drop null values for column 'Android Ver' and column 'Type'
app = app.dropna(subset = ['Type'])
app = app.dropna(axis=0,subset = ['Android Ver'])
app.info()
```

```python
app = app.drop_duplicates(subset='App',keep='first')
```

```python
# rename the columns
app = app.rename(columns={'App':'app','Category':'category','Rating':'rating','Reviews':'reviews','Size':'size','Installs':'installs','Type':'type',
                          'Price':'price','Content Rating':'content_rating','Genres':'genres','Last Updated':'last_updated','Current Ver':'current_ver',
                          'Android Ver':'android_ver'})
app.head(5)
```

```python
app = app.set_index(['app'])
app.head()
```

```python
app.info()
```

```python
# export file
app.to_csv("cleaned_app.csv")

```

# **Reviews Dataset**

```python
reviews = pd.read_csv('/content/drive/MyDrive/Project/playstore_reviews.csv')
reviews.head(3)
```

```python
# remove the duplicates
reviews.drop_duplicates(inplace=True)
reviews.head()
```

```python
print(reviews['Translated_Review'].unique())
print(reviews['Sentiment'].unique())
print(reviews['Sentiment_Polarity'].unique())
print(reviews['Sentiment_Subjectivity'].unique())
```

```python
reviews = reviews.dropna()
reviews.info()
```

```python
reviews = reviews.rename(columns={'App':'app','Translated_Review':'translated_review',
                                  'Sentiment':'sentiment','Sentiment_Polarity':'sentiment_polarity',
                                  'Sentiment_Subjectivity':'sentiment_subjectivity'})
reviews.head()
```

```python
reviews = reviews.set_index(['app'])
reviews.head()
```

```python
reviews.to_csv("cleaned_reviews.csv")
```
